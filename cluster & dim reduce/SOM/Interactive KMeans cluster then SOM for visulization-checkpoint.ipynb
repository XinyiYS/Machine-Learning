```python
import numpy as np
import matplotlib.pylab as plt
import Sachin_som_with_TF as som
import matplotlib, DPP
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
from sklearn.cluster import KMeans
```

```python
dpp = DPP.DPP(data_dir='dataOut.txt')
data, chosen_attributes = dpp.get_data(data_frac = 0.2)

n_features = sum(len(v) for v in chosen_attributes.values())
print('no features is :', n_features)
```

```python
# Cluster the points first before SOM
kmeans = KMeans(n_clusters=4) #try n_cluster = 6
kmeans.fit(data)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Appedning the identifed centroids to the end and train SOM on the entire set
data = np.array(data)
centroids = np.array(centroids)
data_w_centroids = np.concatenate((data,centroids))
```

```python
# initialize the som with size, n_features, and n_itearations
som = som.SOM(15, 15, n_features,  100)
som.train(data_w_centroids)
print("Training is done.")
```

```python
# Map colours to their closest neurons
mapped = np.array(som.map_vects(data_w_centroids))
```

```python
centroids = mapped[-6:]
points = mapped[:-6]
len(points)
len(labels)
```
